```python
import lancedb

# 1. DB가 저장된 경로 설정
uri = "data/vector_db/accidents.lance"

try:
    # DB 연결 시도
    db = lancedb.connect(uri)
    print("✅ 성공: LanceDB에 정상적으로 연결되었습니다.\n")
    
    # 2. 생성된 테이블 목록 확인
    tables = db.table_names()
    print(f"📂 존재하는 테이블 목록: {tables}")
    
    if not tables:
        print("⚠️ 경고: DB는 생성되었으나 안에 테이블이 없습니다.")
    else:
        # 3. 테이블 열기 (첫 번째 테이블을 기준으로 확인하거나, 이름을 직접 지정하세요)
        # 만약 테이블 이름이 'accidents_table' 등으로 다르다면 아래 변수를 수정하세요.
        target_table_name = tables[0] 
        tbl = db.open_table(target_table_name)
        
        print(f"\n▶ [{target_table_name}] 테이블 점검 시작")
        
        # 4. 데이터 개수 확인
        row_count = len(tbl)
        print(f"📊 총 레코드 수: {row_count} 개")
        
        if row_count > 0:
            # 5. 스키마(데이터 구조) 확인
            print("\n구조(Schema):")
            for field in tbl.schema:
                print(f" - {field.name}: {field.type}")
            
            # 6. 샘플 데이터 3개 추출 (Pandas DataFrame 형태로 보기 좋게 출력)
            print("\n🔍 샘플 데이터 (상위 3개):")
            sample_df = tbl.head(3).to_pandas()
            print(sample_df)
        else:
            print("⚠️ 경고: 테이블은 존재하지만 데이터(레코드)가 0개입니다.")

except Exception as e:
    print(f"❌ 오류 발생: DB를 읽는 중 문제가 발생했습니다.\n상세 내용: {e}")
```
